## LINE CHART

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Health_heart_experimental.csv')
sns.lineplot(x='weightKg',y='heightCm',hue='sex',data=df)
```

##### From this line graph it is clear that height vs weight relationship of both male and female is very similar. The trend is linear, indicating that as the weight of a person increases, the height also increases linearly. The height for people with the weight 50 is between 145 to 150. Between 50 t0 60 kg the a minor fluctuation can be observed in the height of male reaching upto 153cm indicating that height of male compared to female at that range is higher.The trend continues until the weight range of 75 to 80 where a small spike can be seen in female height reaching upto 163cm indicating that the height of female at that weight range is somewhat higher than the male poulation, although it too can be considered a minor fluctuation as these differnces are small and do not indicate significant divergence between two sexes. The linear trend then continues with both the sex having similar height vs weight relationship. At the weight of 120 both the male and the female population have nearly the same height of 180cm.

## BAR PLOT

```python
sns.barplot(x='indication',y='BMI',hue='sex',data=df)
```

##### From this bar plot it is evident that the healthier population (INDICATION 0) maintains BMI of approximately 25. Whereas the unhealthy populations (INDICATION 1) have significantly high BMI above 33. There is not much difference between the two sexes and no fluctuations can be seen. This data suggests that a BMI range of 20 to 25 is a strong indication of a healthy person.

## PIE CHART

```python
df['indication'].value_counts().plot(kind='pie',autopct='%0.2f')
```

##### The pie plot suggests that for this specific group, the number of healthy population exceeds the unhelathy population by roughly 5.36%. 

## BOX PLOT 

```python
sns.boxplot(x='sex',y='SysBP',hue='indication',data=df)
```

##### This box plot shows that the unhealthy population (INDICATION 1) have slightly lower median systolic blood pressure as compared to the healthier polulation (INDICATION 0). This may suggest the presence of medications taken by the unhealthy population since the data is contradictary to the scientifically accurate data which shows that the systolic blood pressure for the unhealthy population is generally higher than the healthier population. However it may also suggest the possibility of a medical condition that supresses the blood pressure of the unhealthy population in this group.

```python
sns.boxplot(x='sex',y='DiaBP',hue='indication',data=df)
```

##### For this dataset, the diastolic blood pressure for both the healthy and unhealthy population is evidently equal.

## SCATTER PLOT

```python
sns.scatterplot(x='age', y='BMI',hue='sex', style='indication',data=df)
plt.show()
```

##### From scaterplot, it can be detrmined that for any age, the BMI below 18 or above 30 causes various health risks (INDICATION 1). People with BMI between 19 to 29 do not have any medical conditions (INDICATION 0) so the BMI range between 19 to 29 is found to be healthy. Neither gender is shown to have significant advantage or disadvantage as both groups have same pattern of health risks based strictly on their BMI regardless of their age.
